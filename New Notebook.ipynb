```python
import arcpy
import os
from arcpy.sa import *

b3_input = arcpy.GetParameterAsText(0) 

work_dir = os.path.dirname(b3_input)
arcpy.env.workspace = work_dir
arcpy.env.overwriteOutput = True 

print("Sistem Hazır: Çalışma alanı otomatik olarak ayarlandı!")
```

```python

print("Uydu bantları sisteme yükleniyor...")
b3 = Raster("B3_Mozaik.tif")
b4 = Raster("B4_Mozaik.tif")
b5 = Raster("B5_Mozaik.tif")"B3_Mozaik"

print("NDWI (Su İndeksi) matematiksel olarak hesaplanıyor...")
ndwi_otomatik = (Float(b3) - Float(b5)) / (Float(b3) + Float(b5))
ndwi_otomatik.save("NDWI_Python_2025.tif")

print("NDVI (Kuraklık İndeksi) matematiksel olarak hesaplanıyor...")
ndvi_otomatik = (Float(b5) - Float(b4)) / (Float(b5) + Float(b4))
ndvi_otomatik.save("NDVI_Python_2025.tif")

print("Tebrikler! Tüm analizler Python tarafından saniyeler içinde otomatik olarak üretildi ve klasöre kaydedildi.")
```
